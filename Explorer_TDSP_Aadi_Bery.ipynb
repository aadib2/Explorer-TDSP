**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/Datasets/Motor_Vehicle_Collisions_-_Crashes_20250109.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x = average_crashes_per_hour.index, y= average_crashes_per_hour.values)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Crash Dates')
plt.ylabel('Crashes Per Day')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonal')
decomposition.resid.plot(ax=ax3)
# residual plot allows us to identify irregular events/ unforseen anomalities
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

**Step 3:** Next, we will begin "Severity Mapping." We'll plot crashes on the map and code them based on severity, distinguishing between crashes that resulted in injuries and those that led to fatalities. This will give us a visual representation of where severe crashes tend to occur. </br>

You may initially want to code these incidents by using different colors, but it's important to think about your map and how accessible it is. Will a color-coded map be easy to read for everyone? Let's make a map that's inclusive for people with color blindness by also creating differently-shaped markers (squares, circles, and triangles) for crashes, injuries, and fatalities.

````python
#TODO: Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=42)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
  # red triangle for crashes with fatalities
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "red"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)

  # orange circle for crashes with injuries but no fatalities
    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "orange"  # Injuries
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
  # black squares for crashes with no injuries or fatalities
    else:
        color = "black"  # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color= color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

````python
#TO DO: Begin creating here!
import csv

# Look at residents from each Borough and see the proportion of accidents which had injuries with pedestrians, cyclists, motorists, and deaths with those categories


# split sample data based on boroughs
boroughs = ['BROOKLYN', 'QUEENS', 'MANHATTAN', 'BRONX', 'STATEN ISLAND']

for borough in boroughs:
  borough_subset = data[data['BOROUGH'] == borough]
  crash_types_counts = {
      'Pedestrian Injuries': borough_subset['NUMBER OF PEDESTRIANS INJURED'].sum(),
      'Cyclist Injuries': borough_subset['NUMBER OF CYCLIST INJURED'].sum(),
      'Motorist Injuries': borough_subset['NUMBER OF MOTORIST INJURED'].sum(),
      'Pedestrian Deaths': borough_subset['NUMBER OF PEDESTRIANS KILLED'].sum(),
      'Cyclist Deaths': borough_subset['NUMBER OF CYCLIST KILLED'].sum(),
      'Motorist Deaths': borough_subset['NUMBER OF MOTORIST KILLED'].sum()
  }


  crash_types_df = pd.DataFrame(list(crash_types_counts.items()), columns=['crash Type', 'Count'])

  # Plot
  plt.figure(figsize=(12, 7))
  sns.barplot(x='Count', y='crash Type', data=crash_types_df, palette="mako")
  plt.title(f'Types of crashes and Their Frequencies: {borough}')
  plt.xlabel('Count')
  plt.ylabel('Type of crash')
  plt.tight_layout()
  plt.show()








````

We can visualize this in a map as we did earlier with geospatial analysis.

````python
# Deeper dive into Manhattan
# geospatial analysis
manhattan_data = data_geo[data_geo['BOROUGH'] == 'MANHATTAN']
manhattan_data_sample = manhattan_data.sample(n=2000, random_state=42)

# Create a base map
map_severity = folium.Map(location=[40.7685, -73.9822], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in manhattan_data_sample.iterrows():
  # red triangle for crashes with fatalities
    color = "red"  # Fatalities
    if row['NUMBER OF PERSONS KILLED'] > 0:
      folium.features.RegularPolygonMarker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        number_of_sides=3,
        radius=5,
        gradient = False,
        color=color,
        fill=True,
        fill_color=color
      ).add_to(map_severity)
  # different color circles for crashes with injuries but no fatalities
    elif row['NUMBER OF PERSONS INJURED'] > 0:
        if row['NUMBER OF PEDESTRIANS INJURED'] > 0:
          # orange circles for pedestrian injuries
          color = "orange"
          folium.features.CircleMarker(
            location=[row['LATITUDE'], row['LONGITUDE']],
            radius=5,
            gradient = False,
            color=color,
            fill=True,
            fill_color=color
          ).add_to(map_severity)
        elif row['NUMBER OF CYCLIST INJURED'] > 0:
          # gold circles for cyclist injuries
          color = "gold"
          folium.features.CircleMarker(
            location=[row['LATITUDE'], row['LONGITUDE']],
            radius=5,
            gradient = False,
            color=color,
            fill=True,
            fill_color=color
          ).add_to(map_severity)
        elif row['NUMBER OF MOTORIST INJURED'] > 0:
          # olive green circles for motorist injuries
          color = "olivedrab"
          folium.features.CircleMarker(
            location=[row['LATITUDE'], row['LONGITUDE']],
            radius=5,
            gradient = False,
            color=color,
            fill=True,
            fill_color=color
          ).add_to(map_severity)
  # gray squares for crashes with no injuries or fatalities
    else:
        color = "dimgray" # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color= color,
          fill=True,
          fill_color=color
        ).add_to(map_severity)



map_severity.save("severity_manhattan.html")
````

````python
# remove all entries with no zip code
updated = manhattan_data.dropna(subset=['ZIP CODE'])
# convert zip codes to ints
updated['ZIP CODE'] = updated['ZIP CODE'].astype(int)
updated.info()

# create new dataframe with crash counts for each zip code in Manhattan
crashes_per_zip = updated.groupby('ZIP CODE').size().reset_index(name='CRASH COUNT')
crashes_per_zip.head()
````

Time Series Analysis: Looking at crashes occuring in zip codes with highest concentrations of accidents and see how it has changed over the years and understand why

````python
# identify the top three zip codes have had the greatest concentration of crashes
# create a list of those zip codes with crashes > 14000
top_crashes_per_zip = crashes_per_zip.nlargest(3, 'CRASH COUNT')
display(top_crashes_per_zip)
````

````python
# perform a time series analysis on the crashes occuring in these zip codes

# do some preprocessing
updated['CRASH DATE'] = pd.to_datetime(updated['CRASH DATE'])

#for zip_code in top_crashes_per_zip['CRASH COUNT']:
  # get the subsets from the manhattan_data df
for zip_code in top_crashes_per_zip['ZIP CODE']:
  zip_data = updated[updated['ZIP CODE'] == zip_code]


  zip_monthly_crashes = zip_data.groupby(zip_data['CRASH DATE'].dt.to_period("M")).size()

  # calculate the moving average with our series and display for reduced noise
  moving_average = zip_monthly_crashes.rolling(window = 12).mean()

  sns.set(style = "whitegrid")

  plt.figure(figsize = (15,7))

  # adding labels and title
  zip_monthly_crashes.plot()
  moving_average.plot()
  plt.xlabel('Date')
  plt.ylabel('Monthly Crashes')
  plt.title(f'Number of crashes occuring in zip code #{zip_code} over time')
  plt.tight_layout()
  plt.show()



````

We can see some very noisy data in each of these zip codes as well as many similar trends as we saw earlier due to COVID and seasonal analysis. From 2013 to 2016 there were a lot more crashes occuring on the road with lots of extreme fluctuations. In 2016, we can see a decline in a lot of these crashes but still significant fluctuations. This is after the Vision Zero initiative took off. After the COVID pandemic in 2020 and 2021, there has since been a lot less accidents on the road which we can reasonably infer is due to more people working from home/hybrid, technological advancements in car safety as well as better urban planning with increased bike lanes! There is still a relative seasonal pattern likely due to these areas being popular tourist areas.
